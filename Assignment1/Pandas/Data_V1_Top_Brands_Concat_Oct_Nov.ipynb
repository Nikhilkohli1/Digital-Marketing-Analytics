```python
df_oct = pd.read_csv('2019_Oct_Sales.csv')
```

```python
df_nov = pd.read_csv('2019_Nov_Sales.csv')
```

```python
df_sales = pd.concat([df_oct, df_nov], ignore_index=True)
df_sales.head(10)
```

```python
df_sales['user_session'] = df_sales['user_session'].astype('category').cat.codes.astype('category')
df_sales['user_id'] = df_sales['user_id'].astype('category').cat.codes.astype('category')
df_sales['product_id'] = df_sales['product_id'].astype('category').cat.codes.astype('category')
df_sales['category_id'] = df_sales['category_id'].astype('category').cat.codes.astype('category')
```

```python
df_sales.to_csv('Oct_Nov_2019_Sales.csv')
```

```python
df_sales = df_sales.drop(labels=['Unnamed: 0'], axis=1)
```

```python
#looking at only the cases where purchases were made
purchases = df_sales[df_sales.event_type.eq('purchase')]
purchases
```

```python
top_brands = purchases.groupby('brand')['price'].sum()
top_brands = top_brands.sort_values(ascending= False)
top_brands = top_brands[:50]
```
